<p style="font-weight:bold;">3.1.Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable</p>

```python
#stats of salary_in_usd using describe function
df.salary_in_usd.describe()
```

```python
#sum of all values in salary_in_usd column
sum_of_salary= df.salary_in_usd.sum() 
print(f"The total sum of salary : {sum_of_salary}")
```

```python
#avarage value of salary_in_usd earned
mean_salary = df.salary_in_usd.mean()
print(f"Mean of salary_in_usd : {mean_salary}")
```

```python
#dispersion of dataset relative to its mean
std_salary = df.salary_in_usd.std()
print(f"standard deviation of salary_in_usd : {std_salary}")
```

```python
#to show normal distribution i.e. +ve indicates peaked and -ve indicates flat distribution
kurt_salary = df.salary_in_usd.kurt()
print(f"Kurtosis of salary_in_usd : {kurt_salary}")
```

```python
#0 skewness: distribution is fully symmetrical
#+ve skewness: distribution is skewed to right
#-ve skewness: distribution is skewed to left
skew_salary = df.salary_in_usd.skew()
print(f"Skewness of salary_in_usd : {skew_salary}")
```

<p style="font-weight:bold;">4.1.Write a python program to find out top 15 jobs. Make a bar graph of sales as well.  </p>

```python
# finding top 15jobs
top_15_jobs = df["job_title"].value_counts().head(15)
#finding the job with highest no of people
best_job = top_15_jobs.idxmax()
max_no =  top_15_jobs.max()
print("Top 15 jobs:")
print(top_15_jobs)
print(f"job with highest no : {best_job}")
print(f"max no of jobs in highest job: {max_no}")
```

```python
# plotting in bar graph
x_title = top_15_jobs.index
y_freq = top_15_jobs.values
plt.figure(figsize=(12,6))
plt.bar(x_title,y_freq)
plt.xticks(rotation=90)
plt.title('Top 15 Jobs')
plt.xlabel('Job Title')
plt.ylabel('No of Jobs')
plt.bar(best_job,max_no,color='red') #highlighting the job with highest no
plt.savefig('top_15_jobs.png', bbox_inches='tight')
plt.show()
```

<p style="font-weight:bold;"> 4.2.Which job has the highest salaries? Illustrate with bar graph. </p>

```python
# finding the job with highest salary
salary_by_job =  df.groupby('job_title')['salary_in_usd'].mean()
highest_salaryjob = salary_by_job.idxmax()
max_no_salary = salary_by_job.max()
print("Salaries(USD) of each Job:")
print(salary_by_job)
print(f" \n Job with highest Salary : {highest_salaryjob}")
print(f" \n Max no of salary : {max_no_salary}")
```

```python
# plotting the bar graph
plt.figure(figsize=(12,6))
X_title = salary_by_job.index
Y_salary = salary_by_job.values
plt.bar(X_title,Y_salary)
plt.bar(highest_salaryjob,max_no_salary,color="red")
plt.title(' Salary by Job Title')
plt.xlabel('Job Title')
plt.ylabel(' Salary (USD)')
plt.xticks(rotation=90,fontsize=8)
plt.tight_layout()
plt.savefig('highest_salary.png', bbox_inches='tight')
plt.show()
```

<p style="font-weight:600;color:green;">It can be seen that Data Science Tech Lead has the most paying stipend</p>

<p style="font-weight:bold;"> 4.3.Write a python program to find out salaries based on experience level. Illustrate it through bar graph. </p>

```python
#finding salary based on exp level and the level which gets paid the most on average
salary_on_exp =  df.groupby('experience_level')['salary_in_usd'].mean()
most_paid_salary = salary_on_exp.idxmax()
max_no_sal = salary_on_exp.max()
print("Salary Based On Experience Level:")
print(salary_on_exp)
print(f" \n Most experience level who gets paid : {most_paid_salary}")
print(f"Salary paid : {max_no_salary}")
```

```python
#plotting in bar graph
plt.figure(figsize=(10, 6))
X = salary_on_exp.index
Y = salary_on_exp.values
plt.bar(X, Y)
plt.bar(most_paid_salary,max_no_sal,color="red") #showing the exp level with most paid salary
plt.title('Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('salary_by_expLevel.png', bbox_inches='tight')
plt.show()
```

<p style = "font-weight:600; color:green;">It can be seen that Executive Level gets paid more on average</p>

<p style="font-weight:bold;"> 4.4.Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph. </p>

<h4>Histogram of Salary_in_usd</h4>

```python
min_salary = df['salary_in_usd'].min()
max_salary = df['salary_in_usd'].max()
plt.figure(figsize=(12, 6))
plt.hist(df['salary_in_usd'], bins=40,  color='skyblue', edgecolor='black')
plt.title('Histogram of Salary (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.xticks(np.arange(min_salary, max_salary, (max_salary - min_salary) / 20), rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

<h4>Box Plot of Salary_in_usd and Experience Level</h4>

<p> A Box plot represents statistical five number summary i.e. minimum, Q1, Q2(median),Q3 and maximum  </p>

```python
plt.figure(figsize=(12, 6))
plt.boxplot([df[df['experience_level'] == 'Senior Level']['salary_in_usd'],
             df[df['experience_level'] == 'Intermediate']['salary_in_usd'],
             df[df['experience_level'] == 'Entry Level']['salary_in_usd'],
             df[df['experience_level'] == 'Executive Level']['salary_in_usd']],
            labels=['Senior Level', 'Intermediate', 'Entry Level', 'Executive Level'])
plt.title('Box Plot of Salary by Experience Level')
plt.savefig("boxplot.png")
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

<p style = "color:green; font-size:18px;">The dots present above box/Q3 upper range are known as outliers. it represents the no of people earning more than the one average earns</p>
